```python
import numpy as np
import pandas as pd
import yfinance as yf
import statsmodels.tsa.stattools as ts
import matplotlib.pyplot as plt
import datetime
from scipy.stats import linregress
import colorama
from colorama import Fore, Style
```

```python
def download_data(stock, start, end):
    stock_data = {}
    ticker = yf.download(stock, start, end)
    stock_data['price'] = ticker['Adj Close']
    return pd.DataFrame(stock_data)
```

```python
start_date = datetime.datetime(2019,1,1)
end_date = datetime.datetime(2022,1,1)
```

```python
pair = [download_data("ABFRL.NS",start_date,end_date), download_data("BHEL.NS",start_date,end_date), download_data("CANBK.NS",start_date,end_date), download_data("EXIDEIND.NS",start_date,end_date), download_data("GMRINFRA.NS",start_date,end_date), download_data("ISEC.NS",start_date,end_date), download_data("JINDALSTEL.NS",start_date,end_date), download_data("PFIZER.NS",start_date,end_date), download_data("RAYMOND.NS",start_date,end_date), download_data("ZEEL.NS",start_date,end_date)]
stock_names = ["ABFRL.NS","BHEL.NS","CANBK.NS","EXIDEIND.NS","GMRINFRA.NS","ISEC.NS","JINDALSTEL.NS","PFIZER.NS","RAYMOND.NS","ZEEL.NS"]
```

```python
for i in range(10):
    for j in range(10):
        result = linregress(pair[i].values[:,0], pair[j].values[:,0])
        
        residuals = pair[i] - result.slope*pair[j]
        adf = ts.adfuller(residuals)
        
        print(stock_names[i],stock_names[j])
        print(result)
        print("AD Fuller Test :")
        print(adf)
        
        if(adf[0]<-2.8655458544300387):
            print(Fore.RED + 'Pair has stationary spread')
            print(Style.RESET_ALL)
        print("\n")
        
```
